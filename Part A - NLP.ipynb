# Stop Words

# Lemmatization
Lemmatization is the process of reducing inflected forms of a word while still ensuring that the reduced form belongs to the language. This reduced form, or root word, is called a lemma.

For example, `organizes`, `organized` and `organizing` are all forms of `organize`. Here, organize is the lemma. The inflection of a word allows you to express different grammatical categories, like tense (`organized` vs `organize`), number (`trains` vs `train`), and so on. Lemmatization is necessary because it helps you reduce the inflected forms of a word so that they can be analyzed as a single item. It can also help you normalize the text.

spaCy puts a `lemma_` attribute on the Token class. This attribute has the lemmatized form of the token.

```python
# Sample Text
text = (
    "He ate an apple."
    "I saw eighteen wolves yesterday!"
)

# Create a Doc object
doc = nlp(text)
```

# Part-of-Speech Tagging

**Part of speech** or **POS** is a grammatical role that explains how a particular word is used in a sentence. There are typically eight main parts of speech:

1. Noun
2. Pronoun
3. Adjective
4. Verb
5. Adverb
6. Preposition
7. Conjunction
8. Interjection

Part-of-speech tagging is the process of assigning a POS tag to each token depending on its usage in the sentence. POS tags are useful for assigning a syntactic category like noun or verb to each word.

In spaCy, POS tags are available as an attribute on the Token object.

* To view the POS tag use `token.pos_`
* To view the fine-grained tag use `token.tag_`
* To view the description of either type of tag use `spacy.explain(tag)`

You may visit [this link](https://v2.spacy.io/api/annotation#pos-tagging) to get the complete list of tags.

# Visualizing Parts of Speech

You can use **displaCy** to find POS tags for tokens.

The dependency parse shows the POS tag for each token, as well as the **dependency tag** if given.

* To view the dependency tag use `token.dep_`
* To view the description of either type of dependency use `spacy.explain(dep)`

```python
print(
    f'{"Token":15}'
    f'{"POS":8}'
    f'{"Dependency":15}'
    f'{"Description"}'
)
print("-"*50)
for token in doc:
    print(
        f"{str(token.text):15}"
        f"{str(token.pos_):8}"
        f"{str(token.dep_):15}"
        f"{str(spacy.explain(token.dep_))}"
    )
```

# Named-Entity Recognition
**Named-entity recognition** (**NER**) is the process of locating named entities in unstructured text and then classifying them into predefined categories, such as person names, organizations, locations, monetary values, percentages, and time expressions.

You can use **NER** to learn more about the meaning of your text. For example, you could use it to populate tags for a set of documents in order to improve the keyword search. You could also use it to categorize customer support tickets into relevant categories.

spaCy has the property `.ents` on Doc objects. You can use it to extract named entities.

* To view the NER text use `ent.text`
* To view the NER label use `ent.label_`
* To view the description of either type of NER use `spacy.explain(ent.label_)`

# Rule-Based Matching Using spaCy
**Rule-based matching** is one of the steps in extracting information from unstructured text. It‚Äôs used to identify and extract tokens and phrases according to patterns (such as lowercase) and grammatical features (such as part of speech).

While you can use regular expressions to extract entities (such as phone numbers), rule-based matching in spaCy is more powerful than regex alone, because you can include semantic or grammatical filters.

‚ùì <font color='red'>Question: </font> For example, with rule-based matching, you can extract a first name and a last name, which are always proper nouns. Complete the following code to extract first name and a last name.

```python
from spacy.matcher import Matcher
matcher = Matcher(nlp.vocab)

# Sample Text
text = "John Doe completed his PhD degree in computer science in 2016."

# Create a Doc object
doc = nlp(text)

pattern = [{"POS": _____}, {"POS": _____}]
matcher.add("FULL_NAME", [pattern])
matches = matcher(doc)
for _, start, end in matches:
    span = doc[start:end]
    print(span.text)
```

The **PhraseMatcher** lets you efficiently match large terminology lists. While the Matcher lets you match sequences based on lists of token descriptions, the PhraseMatcher accepts match patterns in the form of Doc objects.

```python
from spacy.matcher import PhraseMatcher
matcher = PhraseMatcher(nlp.vocab, attr='LOWER')

terms = ['AI', 'coursework', 'Chatbot', 'Train' , 'ticket', 'delay']

# create pattern objects from text terms
patterns = [nlp(text) for text in terms]

# add the patterns to a term list
matcher.add("TermList", patterns)

# Sample Text
text = "The AI coursework is to develop an intelligent Chatbot for finding the cheapest train ticket and predicting train delay."

# Create a Doc object
doc = nlp(text)

#match the text to the patterns
matches = matcher(doc)

print(
    f'{"Matched Pattern":20}'
    f'{"Start Position":20}'
    f'{"End Position"}'
)

for match_id, start, end in matches:
      print(
        f"{str(doc[start:end]):20}"
        f"{str(start):20}"
        f"{str(end)}"
    )
```
